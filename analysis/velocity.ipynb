# Velocity checker
Script checking if the cars move backwards

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../build/data/multi_lane.csv', index_col=False)
```

```python
df.head()
```

```python
data = df.to_numpy()

t = data[:,0]
t_shortened = t[:-1]
car_data = data[:, 1:]
location = car_data[:, ::2]
lane = car_data[:, 1::2]
```

```python
for i in range(10,20):
    string = 'x' + str(i)
    x = df[string]
    dt_array = np.ones(len(np.diff(x))) * t[1]
    v = np.diff(x) / dt_array

    plt.ylim(-2, 30)

    plt.plot(t_shortened, v, label='Car ' + str(i))
    
plt.legend()
plt.tight_layout()
```
